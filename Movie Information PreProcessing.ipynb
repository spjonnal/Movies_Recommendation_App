```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
final = pd.read_csv('Final Movie Dataset.csv')
movies = pd.read_csv('movies_final_data.csv')
```

```python
warnings.filterwarnings('ignore')
for i in range(movies.shape[0]):
    duration = movies['runtime'].iloc[i]

    if pd.isna(duration):
        movies['runtime'].iloc[i] = 'NaN'
        continue
    else:
        # Ensure the duration is numeric
        try:
            time = int(duration)
            hours = time // 60
            minutes = time % 60
            movies['runtime'].iloc[i] = f"{hours}h {minutes}m"
        except:
            movies['runtime'].iloc[i] = 'NaN'
```

```python
movies= movies.drop(columns=['genres','id','vote_count','tagline','original_title','original_language','spoken_languages','popularity'])
```

```python
movies['genres_included'] = movies['genres_included'].str.split(',')
movies = movies.explode('genres_included').reset_index(drop=True)
movies.head()
```

```python
movies = movies.rename(columns={'vote_average':'Ratings','cast_info_complete':'Cast and Crew','imdb_id':'IMDB ID','adult':'Is Adult Rated?','overview':'Overview','release_date':'Release Data',
                                'runtime':'Runtime','title':'Title','genres_included':'Genre','available_languages':'Available Languages'})
movies.head()
```

```python
movies.to_csv('final_movie_data.csv',index=False)
```
